# Scaling

```python
#using fit.transform
x = X_train.values #returns a numpy array
test_s = test.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled_train = min_max_scaler.fit_transform(x)
x_scaled_test = min_max_scaler.fit_transform(test_s)
df_train = pd.DataFrame(x_scaled_train)
df_test = pd.DataFrame(x_scaled_test)
```

```python
#visualise top 5 rows
df_train.head()
```

# Building the model

```python
#splitting the datas
X_train, X_test, y_train, y_test = train_test_split(df_train, y_train, test_size=0.4)
```

```python
#using linear regression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
#scatterplot
plt.scatter(y_test, preds)
plt.show()
```

```python
#distribution plot
sns.distplot((y_test-preds),bins=50)
plt.show()
```

# Model Evaluation

```python
#mean_absolute_error, mean_squared_error, mean_squared_error
print("MAE:", metrics.mean_absolute_error(y_test, preds))
print('MSE:', metrics.mean_squared_error(y_test, preds))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, preds)))
```

```python
x1=Bdr.iloc[:,0:-1]
x1.head()
```

```python
y=Bdr.iloc[:,-1]
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x1)
x=pd.DataFrame(x,columns=x1.columns)
```

```python
x.shape
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
max_r_score=0
for r_state in range(40,3000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=.20)
    regr=linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred=regr.predict(x_test)
    r_scr=r2_score(y_test,y_pred)
    if r_scr>max_r_score:
        max_r_score=r_scr
        final_r_state=r_state
print('max r2 score corresponding to',final_r_state,'is',max_r_score)
```

```python
from sklearn.model_selection import cross_val_score
a_score=cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring='r2')
a_score
```
